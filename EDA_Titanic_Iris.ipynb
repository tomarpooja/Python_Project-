```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.histplot(titanic["age"], bins=20, kde=True)
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

```python
sns.boxplot(x="class", y="age", data=titanic)
plt.title("Age Distribution per Class")
plt.show()
```

```python
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.shape          # Rows × Columns
iris.info()         # Data types + null values
iris.describe()     # Summary stats
iris.isnull().sum() # Check missing values
```

```python
sns.histplot(iris['sepal_length'], kde=True, bins=20)
plt.title("Sepal Length Distribution")
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title("Petal Length per Species")
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
